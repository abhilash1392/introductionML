```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 
import mglearn
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X,y=mglearn.datasets.make_wave(n_samples=100)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
print(f'Coefficient: {reg.coef_}')
print(f'Intercept: {reg.intercept_}')
print(f'Training Accuracy: {reg.score(X_train,y_train):.2f}')
print(f'Testing Accuracy: {reg.score(X_test,y_test):.2f}')
```

```python
# Fitting it on the boston housing extended database
X,y=mglearn.datasets.load_extended_boston()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
lr=LinearRegression().fit(X_train,y_train)
print(f'Training Score: {lr.score(X_train,y_train):.2f}')
print(f'Testing Score: {lr.score(X_test,y_test):.2f}')
```

```python
# Ridge Regression
from sklearn.linear_model import Ridge
ridge=Ridge().fit(X_train,y_train)
print(f'Training Score: {ridge.score(X_train,y_train):.2f}')
print(f'Testing Score: {ridge.score(X_test,y_test):.2f}')
```

```python
ridge10=Ridge(alpha=10).fit(X_train,y_train)
print(f'Training Score: {ridge10.score(X_train,y_train):.2f}')
print(f'Testing Score: {ridge10.score(X_test,y_test):.2f}')
```

```python
ridge01=Ridge(alpha=0.1).fit(X_train,y_train)
print(f'Training Score: {ridge01.score(X_train,y_train):.2f}')
print(f'Testing Score: {ridge01.score(X_test,y_test):.2f}')
```

```python
plt.plot(ridge.coef_,marker='s',label="Ridge Alpha 1",lw=0)
plt.plot(ridge10.coef_,marker='o',label="Ridge Alpha 10",lw=0)
plt.plot(ridge01.coef_,marker='x',label="Ridge Alpha 0.1",lw=0)
# plt.plot(lr.coef_,marker='^',label='Linear Regression')
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.hlines(0, 0, len(lr.coef_))

plt.legend()
plt.show()
```

```python
# Lasso
from sklearn.linear_model import Lasso
lasso=Lasso().fit(X_train,y_train)
print(f'Training Score: {lasso.score(X_train,y_train):.2f}')
print(f'Testing Score: {lasso.score(X_test,y_test):.2f}')
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))
```

```python
lasso001=Lasso(alpha=0.01,max_iter=100000).fit(X_train,y_train)
print(f'Training Score: {lasso001.score(X_train,y_train):.2f}')
print(f'Testing Score: {lasso001.score(X_test,y_test):.2f}')
print("Number of features used: {}".format(np.sum(lasso001.coef_ != 0)))
```

```python
lasso00001=Lasso(alpha=0.0001,max_iter=100000).fit(X_train,y_train)
print(f'Training Score: {lasso00001.score(X_train,y_train):.2f}')
print(f'Testing Score: {lasso00001.score(X_test,y_test):.2f}')
print("Number of features used: {}".format(np.sum(lasso00001.coef_ != 0)))
```

```python
plt.plot(lasso.coef_,marker='v',label='Alpha 1',lw=0)
plt.plot(lasso001.coef_,marker='x',label='Alpha 0.01',lw=0)
plt.plot(lasso00001.coef_,marker='^',label='Alpha 0.0001',lw=0)
plt.ylim(-25,25)
plt.legend()
plt.hlines(0,0,len(lr.coef_),colors='black')
plt.xlabel('Coefficient Index')
plt.ylabel('Coefficient Magnitude')
plt.show()
```
